```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import pandas as pd
import seaborn

seaborn.set(style="ticks")
seaborn.set(font_scale=1.2)  
seaborn.set_style("darkgrid")
```

```python
df = pd.read_json("simulation_data.json")
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

df.set_index("Basis degree").loc[[3]].pivot(index='Degrees of freedom', columns='CFL', values='L2 error (FE)')[0.1125].plot()
df.set_index("Basis degree").loc[[3]].pivot(index='Degrees of freedom', columns='CFL', values='L2 error (BA)')[0.1125].plot(style='--')

df.set_index("Basis degree").loc[[2]].pivot(index='Degrees of freedom', columns='CFL', values='L2 error (FE)')[0.9].plot()
df.set_index("Basis degree").loc[[2]].pivot(index='Degrees of freedom', columns='CFL', values='L2 error (BA)')[0.9].plot(style='--')

df.set_index("Basis degree").loc[[4]].pivot(index='Degrees of freedom', columns='CFL', values='L2 error (FE)')[0.9].plot()
df.set_index("Basis degree").loc[[4]].pivot(index='Degrees of freedom', columns='CFL', values='L2 error (BA)')[0.9].plot(style='--')

ax.set_xscale('log')
ax.set_yscale('log')
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

seaborn.heatmap(df.set_index("Basis degree").loc[[5]].pivot(index='Degrees of freedom', columns='CFL', values='L2 error (FE)'), ax=ax, square=True, annot=True, norm=LogNorm())
```
