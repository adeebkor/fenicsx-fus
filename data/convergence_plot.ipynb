```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import pandas as pd
import seaborn

seaborn.set(style="ticks")
seaborn.set(font_scale=1.2)  
seaborn.set_style("darkgrid")
```

```python
df = pd.read_json("simulation_data.json")
df
```

```python
df.set_index("Basis degree").loc[4]
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

seaborn.heatmap(df.set_index("Basis degree").loc[5].pivot(index='Degrees of freedom', columns='Tolerance', values='H1 error (FE)'), ax=ax, square=True, annot=True, norm=LogNorm())
```
