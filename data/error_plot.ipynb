```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
# Plotting setup
font = {'size': 18, 'family': 'sans-serif'}

rc('font', **font)
# rc('text', usetex=True)
```

```python
# 1D linear wave equation
linear_p1 = pd.read_csv("error_linear_1d_p1.csv", names=["Number of element per wavelength", "Degree of freedoms", "L2 error"], skiprows=1)
linear_p2 = pd.read_csv("error_linear_1d_p2.csv", names=["Number of element per wavelength", "Degree of freedoms", "L2 error"], skiprows=1)
linear_p3 = pd.read_csv("error_linear_1d_p3.csv", names=["Number of element per wavelength", "Degree of freedoms", "L2 error"], skiprows=1)
linear_p4 = pd.read_csv("error_linear_1d_p4.csv", names=["Number of element per wavelength", "Degree of freedoms", "L2 error"], skiprows=1)
```

```python
linear_p4
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(linear_p1["Degree of freedoms"], linear_p1["L2 error"], 'C0-o', ms=5, label='$p=1$')
ax.plot(linear_p2["Degree of freedoms"], linear_p2["L2 error"], 'C1-o', ms=5, label='$p=2$')
ax.plot(linear_p3["Degree of freedoms"], linear_p3["L2 error"], 'C2-o', ms=5, label='$p=3$')
ax.plot(linear_p4["Degree of freedoms"], linear_p4["L2 error"], 'C3-o', ms=5, label='$p=4$')


ax.set_xscale('log')
ax.set_yscale('log')

ax.set_title('Linear wave equation')
ax.set_xlabel('Degree of freedoms')
ax.set_ylabel('$|u-u_{h}|/|u|$')

plt.grid('on', which='both', linestyle='--')
plt.legend()
```

```python
# 1D Westervelt equation
westervelt_p1 = pd.read_csv("error_westervelt_1d_p1.csv", names=["Number of element per wavelength", "Degree of freedoms", "L2 error"], skiprows=1)
westervelt_p2 = pd.read_csv("error_westervelt_1d_p2.csv", names=["Number of element per wavelength", "Degree of freedoms", "L2 error"], skiprows=1)
westervelt_p3 = pd.read_csv("error_westervelt_1d_p3.csv", names=["Number of element per wavelength", "Degree of freedoms", "L2 error"], skiprows=1)
westervelt_p4 = pd.read_csv("error_westervelt_1d_p4.csv", names=["Number of element per wavelength", "Degree of freedoms", "L2 error"], skiprows=1)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(westervelt_p1["Degree of freedoms"], westervelt_p1["L2 error"], 'C0-o', ms=5, label='$p=1$')
ax.plot(westervelt_p2["Degree of freedoms"], westervelt_p2["L2 error"], 'C1-o', ms=5, label='$p=2$')
ax.plot(westervelt_p3["Degree of freedoms"], westervelt_p3["L2 error"], 'C2-o', ms=5, label='$p=3$')
ax.plot(westervelt_p4["Degree of freedoms"], westervelt_p4["L2 error"], 'C3-o', ms=5, label='$p=4$')


ax.set_xscale('log')
ax.set_yscale('log')

ax.set_title('Westervelt equation')
ax.set_xlabel('Degree of freedoms')
ax.set_ylabel('$|u-u_{h}|/|u|$')

plt.grid('on', which='both', linestyle='--')
plt.legend()
```
